```python
import time
import random
from city_plot import City, CityPlot
```

```python
def random_cities(n, X1, X2, Y1, Y2):
    cities = []
    for i_ in range(n):
        x_, y_ = random.uniform(X1 / 2, X2 / 2), random.uniform(Y1, Y2)
        city = City(i = i_, x = x_, y = y_)
        cities.append(city)
    return cities
```

```python
NUMBER_OF_CITIES = 10
X1, X2, Y1, Y2 = 0.0, 20.0, -6.0, 6.0 # Range of plot 

cities = random_cities(NUMBER_OF_CITIES, X1, X2, Y1, Y2)
plot = CityPlot(cities, (X1, X2), (Y1, Y2))
plot.show()

order1, order2 = [i for i in range(0, NUMBER_OF_CITIES)], [i for i in range(0, NUMBER_OF_CITIES)]

for x in range(100):
    time.sleep(0.1)
    random.shuffle(order1)
    random.shuffle(order2)
    plot.update(order1, order2)
    
print("done!")
```
